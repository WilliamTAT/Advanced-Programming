```python
from nltk.corpus import PlaintextCorpusReader as PCR
from nltk import word_tokenize, Text, FreqDist
```

```python
#该类的一个实例
files = PCR("C:/Users/Aimee/corpora",'.*\.txt') #/方向
```

```python
#符合格式的文件的所有文件名
files.fileids()
```

```python
help(word_tokenize)
```

```python
##把所有符合格式的文件都读进来
#raw方法将所有字符串抽出来合并到一块
raw_text = files.raw(fileids=files.fileids())
```

```python
type(raw_text)
```

```python
#word_tokenize按空格、标点等模型切分
#词袋子，word bag
words = word_tokenize(raw_text)
```

```python
#Text是生成语料库的类
cps1 = Text(words)
```

```python
dir(cps1)
```

```python
fdist = FreqDist(cps1)
```

```python
#概率分布
type(fdist)
```

```python
w1 = fdist.items()
```

```python
#按第二个关键字倒序排序
w1_sorted_desc = sorted(w1, key=lambda w:w[1], reverse = True)
w1_sorted_desc
```
